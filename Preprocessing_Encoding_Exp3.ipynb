# Dropping Rows with Null Values

```python
# Dropping Rows with null Values
df = df.dropna()
```

# ***Applying Different Encoding Techniques***

# **1. Label Encoding for Binary Categorical Columns**

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
```

```python
df['Gender'] = label_encoder.fit_transform(df['Gender'])
```

```python
df.head()
```

# **2. One-Hot Encoding for Multi-Class Categorical**

```python
df = pd.get_dummies(df, columns=['cuisine'], drop_first=True)
```

```python
df.head()
```

# **3. Ordinal Encoding for Ordered Categorical Columns**

```python
from sklearn.preprocessing import OrdinalEncoder
ordinal_encoder = OrdinalEncoder(categories=[['low', 'medium', 'high']])
```

```python
df['healthy_feeling'] = df['healthy_feeling'].astype(str)
```

```python
# Determine categories based on unique values
unique_values = df['healthy_feeling'].unique()
categories = sorted(set(unique_values))  # Create a sorted list of unique values

# Initialize OrdinalEncoder with the updated categories
ordinal_encoder = OrdinalEncoder(categories=[categories], handle_unknown='use_encoded_value', unknown_value=-1)

# Apply Ordinal Encoding
df['healthy_feeling'] = ordinal_encoder.fit_transform(df[['healthy_feeling']])
```

```python
df = pd.get_dummies(df, columns=['healthy_feeling'])
```
